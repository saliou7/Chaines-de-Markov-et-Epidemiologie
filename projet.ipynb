```python


#Code de la fonction est_stochastique : 
def est_stochastique(matrice) :
    
    nb_ligne=len(matrice)
    nb_colonne=len(matrice[0])
    
    """on va réaliser une boucle qui calcule la somme des valeurs de chaque ligne, si une case
    est >1, alors fonction renvoie faux, si la somme de la ligne est >1, alors faux aussi, si on parcourt toute la
    matrice sans rencontrer de problème, alors la fonction retourne vrai.
    """

    for i in range(nb_ligne): 
        somme=0
        for j in range(nb_colonne):
            if matrice[i][j] <= 1 and matrice[i][j] >= 0:
                somme+=matrice[i][j]
            else: 
                return False
        if somme != 1:
            return False 
    return True


#test avec la matrice de la partie 1 :            
print(est_stochastique(res)) 
#test avec une matrice non stochastique
Mat=[[0,0.2,0.8],[0.6,0.9,0.1]]
print(est_stochastique(Mat))



```

3. De même pour chaque temps t entre 1 et 200, calculez la distribution théorique des effectifs dans
chaque état (Rappel πt+1 = πtA )

```python
"""on réalise une fonction qui va renvoyer le tableau qui contiendra les distributions théoriques de chaque
état pour chaque temps entre 1 et n, avec n=200 ici. La 1ere colonne représentera l'état sain, la deuxième
l'état infecté et la troisième l'état guéri.
""" 
def distrib_1(n,etat_init,mat_trans):  
    
    Mat = np.zeros((n,3))
    # cas de base pour t=1
    
    #on initialise la première ligne grâce à la formule qu'on a déduite des calculs précédents
    Mat[0][0]+=mat_trans[0][0]*etat_init[0]+mat_trans[2][0]*etat_init[2]
    Mat[0][1]+=mat_trans[0][1]*etat_init[0]+mat_trans[1][1]*etat_init[1]
    Mat[0][2]+=mat_trans[1][2]*etat_init[1]+mat_trans[0][2]*etat_init[2]
    
    """on cherche à chaque fois dans les matrices données en arguments au lieu de prendre directement 
    les valeurs de A et pi0 pour pouvoir généraliser la fonction à d'autres matrices de transitions
    et d'autres états initiaux"""
    
    for i in range(1,n):
        for j in range(3):
            if j==0:
                Mat[i][j]=mat_trans[0][0]*Mat[i-1][j]+mat_trans[2][0]*Mat[i-1][j+2]
            else:
                if j==1:
                    Mat[i][j]=mat_trans[0][1]*Mat[i-1][j-1]+mat_trans[1][1]*Mat[i-1][j]
                else:
                    Mat[i][j]=mat_trans[1][2]*Mat[i-1][j-1]+mat_trans[2][2]*Mat[i-1][j]
    return Mat

#test 
distrib_1(200,pi0,A)
```

```python
"""
on réalise une fonction qui va renvoyer le tableau qui contiendra les distributions théoriques de chaque
état pour chaque temps entre 1 et n, avec n=200 ici. La 1ere colonne représentera l'état sain, la deuxième
l'état infecté et la troisième l'état guéri.
"""

#    Nous allons utiliser la formule " pi_k = pi_0 x A^k " pour calculer la probabilité des états au temps k

def distrib1(n,etat_init,mat_trans):
    res = np.array([])
    for i in range (1, n+1) : 
        m_transition = np.linalg.matrix_power(mat_trans, i) # mat_trans a la puissance i
        res = np.append(res, np.dot(etat_init, m_transition)) # multiplier pi0 par A^n
    return np.reshape(res, (n, 3))  # transformer en forme array pour que ça soit plus lisible
#test
distrib1(200,pi0,A)
```

<strong>Tirage aléatoire des états

Vous allez générer une séquence de taille T en utilisant cette chaîne de Markov. Pour générer une
séquence aléatoire, choisissiez un état initial au hasard (en utilisant π0) ; puis choisissez les états suivants
en suivant les probabilités de transition (= la matrice de transition A).
Vous pouvez prendre T=150.

```python
#la fonction prend 3 arguments : t la taille de la séquence générée, etat_init qui correspond ici à pi0 (énoncé), et 
#mat_trans qui est la matrice des probabilités de transition, ici la matrice A
def seq_alea(t,etat_init,mat_trans):

    seq=[] 

#on génère un état initial aléatoire 
    rand=rd.uniform(0,1) # rand est un nombre aléatoire compris entre 0 et 1
    if rand<=etat_init[0]:
        seq.append(0)
    elif rand<=etat_init[0]+etat_init[1]:
        seq.append(1)
    else:
        seq.append(2)
        
    """Maintenant qu'on a notre état de base, on peut générer la séquence. Pour ce faire, on prend un random entre 0 et 1,
si rand est inférieur à la proba d'être sain sachant notre état du rang précédent, alors on ajoute un sain à la séquence,
sinon, si il est entre la proba d'être sain sachant notre état du rang précédent et la proba d'être infecté sachant
notre état du rang précédent, on ajoute un infecté à la séquence, sinon on rajoute un guéri à la séquence.
"""   
    for i in range(t-1):
        rand=rd.uniform(0,1) 
        if rand<=mat_trans[seq[-1]][0]:
            seq.append(0)
        else:
            if rand<=mat_trans[seq[-1]][0] + mat_trans[seq[-1]][1]:
                seq.append(1)
            else:
                seq.append(2)
    return seq

#on test avec la matrice de transitions A
A=np.array([[0.92,0.08,0],[0,0.93,0.07],[0,0,1]])
print(seq_alea(150,pi0,A))     
```

```python
#on code la fonction nb_individus qui affiche le nombre d'individus sains infecté et guéri dans la population
#à chaque temps ti jusqu'au temps tn. 
def nb_individus(n,m,etat_init,mat_trans): # n : nb ité; m : pop
    
    etat_S=np.zeros(n)
    etat_I=np.zeros(n)
    etat_R=np.zeros(n)
    
    """la boucle va créer m seq aléatoire, et on va regarder pour chaque temps j le nombre total parmi les
    m individus de chaque état"""
    for i in range(m):
        etat=seq_alea(n,etat_init,mat_trans)
        for j in range (n):
            if etat[j]==0:
                etat_S[j]+=1
            elif etat[j]==1:
                etat_I[j]+=1
            else:
                etat_R[j]+=1
    res=[etat_S,etat_I,etat_R]
    return res

# test 
etat_S,etat_I,etat_R = nb_individus(150,5,pi0,A)
print(etat_S)
print(etat_I)
print(etat_R)
```

2. Affichez le pourcentage d’indivudus sains infectés et guéris en fonction du temps.


```python
# on réalise une fonction quik donne le graphe en pourcentage 
def graphe_pop_pourc(n,pop,etat_init,mat_trans): # n -> nb ité  ;  pop ->taille population
    t=[]
    nb_S=[] # représente nombre de sains dans le graphique 
    nb_I=[]
    nb_R=[]
    res=nb_individus(n,pop,etat_init,mat_trans)

    for i in range(n):
        nb_S.append(res[0][i]/float(pop))
        nb_I.append(res[1][i]/float(pop))
        nb_R.append(res[2][i]/float(pop))
        t.append(i)

    #on créé le graphique 
    fig, ax = plt.subplots()
    ax.plot(t,nb_S,label="Nb sains")
    ax.plot(t,nb_I,label="Nb_infectes")
    ax.plot(t,nb_R,label="Nb_gueris")
    ax.set_xlabel("Ti")
    ax.set_ylabel("Nombre")
    ax.legend()
    plt.show

#affichage du graphique 
print(graphe_pop_pourc(150,200,pi0,A))
```

```python
#affichage d'un graphique suivant l'évolution du nombre de personnes saines, infectés et guéries de l'itération 1 à n
def graphe_pop(n,pop,etat_init,mat_trans): # n -> nb ité  ;  pop ->taille population
    t=[]
    nb_S=[] # représente nombre de sains dans le graphique 
    nb_I=[]
    nb_R=[]
    res=nb_individus(n,pop,etat_init,mat_trans)

    for i in range(n):
        nb_S.append(res[0][i])
        nb_I.append(res[1][i])
        nb_R.append(res[2][i])
        t.append(i)

    #on créé le graphique 
    fig, ax = plt.subplots()
    ax.plot(t,nb_S,label="Nb sains")
    ax.plot(t,nb_I,label="Nb_infectes")
    ax.plot(t,nb_R,label="Nb_gueris")
    ax.set_xlabel("Ti")
    ax.set_ylabel("Nombre")
    ax.legend()
    plt.show
# exec de la fonction
print(graphe_pop(150,200,pi0,A))
    
```

```python
"""n=nb itérations
m= taille de la population
e désigne l'etat dont on recherche la longueu moyenne. Le 0 correspond à l'état sain, le 1 à l'état infecté et le 2 à
l'état guéri. Cete variable supplémentaire permet de généraliser la fonction, dans le cas ou nous voudrions calculer la 
longueur moyenne d'une autre séquence qu'infecté"""
def long_moy_etat(n,m,etat_init,mat_trans,e):
    
    res=[None]*m
    cpt=0
    
    for i in range(m):
        etat=seq_alea(n,etat_init,mat_trans)
        for j in range(n):
            if etat[j]==e:
                cpt+=1
        res[i]=cpt
        cpt=0
    
    somme=0
    for k in range(m):
        somme+=res[k]
        
    print("\nlongueur moyenne d'une séquence I : ",somme/m)
    return res
# test 
long_moy=long_moy_etat(150,200,pi0,A,1)
#print(long_moy)
              
```

```python
def sim_virus(ite,pop,etat_init,mat_trans):
    
    
    
    #tableau qui recense toute la population,chaque personne a un état,si la ieme personne est saine, alors pop_tab[i]=0
    #si elle est infectée, alors =1 et si guérie =2
    pop_tab=[0]*pop 
    
    # nombre de personne dans chaque état qui nous permettra de calculer le poucentage e personnes infectées 
    nb_sain=[0]*ite
    nb_sain[0] = pop
    nb_inf=[0]*ite
    nb_gueri=[0]*ite

    #matrice de transitions en période de confinement, faux c'est l'etat qu'il faut changer
    mat_confinement= mat_trans.copy()
    mat_confinement[0][0]=1
    mat_confinement[0][1]=0
    
 
    """ à chaque itération, on est soit en confinement soit en période libre. On fait evoluer l'etat de chaque membre
    de la population selon les matrices de transitions hors et en confinement, si un confinement ou une fin de confinement 
    est nécessaire, on change la valeur du booleen"""
    
    confinement = False
    i=1
    while(i < ite):
        for j in range(0,pop):
            rand=rd.uniform(0,1)
            
            if not confinement :  # pas de confinement  
                if pop_tab[j]==0: # cas ou la personne étudiée était saine
                    if rand>mat_trans[0][0]:
                        nb_inf[i]+=1
                        pop_tab[j]=1
                    else :
                        nb_sain[i]+=1
                        
                elif pop_tab[j]==1: # cas ou la personne étduiée était infectée
                    if rand>mat_trans[1][1]:
                        nb_gueri[i]+=1
                        pop_tab[j]=2
                    else: 
                        nb_inf[i]+=1
                else:              # cas ou la personne e=étduiée était guérie
                    if rand>mat_trans[2][2]:
                        nb_sain[i]+=1
                        pop_tab[j]=1
                    else:
                        nb_gueri[i]+=1
            
            else: # cas dans lequel on est en periode de confinement
                if pop_tab[j]==0: # cas ou la personne étudiée était saine
                    if rand>mat_confinement[0][0]:
                        nb_inf[i]+=1
                        pop_tab[j]=1
                    else :
                        nb_sain[i]+=1
                        
                elif pop_tab[j]==1: # cas ou la personne étduiée était infectée
                    if rand>mat_confinement[1][1]:
                        nb_gueri[i]+=1
                        pop_tab[j]=2
                    else: 
                        nb_inf[i]+=1
                else:              # cas ou la personne e=étduiée était guérie
                    if rand>mat_confinement[2][2]:
                        nb_sain[i]+=1
                        pop_tab[j]=1
                    else:
                        nb_gueri[i]+=1
                        
        if nb_inf[i]<((10*pop)/100): #si la population infectée est en dessous de 10% 
            confinement = False
        if nb_inf[i]>((25*pop)/100): #si la population infectée depasse 25%
            confinement = True
        i+=1
    return [nb_sain,nb_inf,nb_gueri]
                

nb_sain,nb_inf,nb_gueri = sim_virus(100,5000,pi0,A)
print(nb_sain)
print(nb_inf)
print(nb_gueri)
print(A2)
#print(sim_virus(150,5000,pi0,A2))
```

```python
# on réalise une fonction qui nous affichera le graphique correspondant à la fonction sim_virus
def graphe_pop_pourc(ite,pop,etat_init,mat_trans): # n -> nb ité  ;  pop ->taille population
    t=[]
    nb_S=[] # représente nombre de sains dans le graphique 
    nb_I=[]
    nb_R=[]
    res=sim_virus(ite,pop,etat_init,mat_trans)

    for i in range(ite):
        nb_S.append(res[0][i]/float(pop))
        nb_I.append(res[1][i]/float(pop))
        nb_R.append(res[2][i]/float(pop))
        t.append(i)

    #on créé le graphique 
    fig, ax = plt.subplots()
    ax.plot(t,nb_S,label="Nb sains")
    ax.plot(t,nb_I,label="Nb_infectes")
    ax.plot(t,nb_R,label="Nb_gueris")
    ax.set_xlabel("Ti")
    ax.set_ylabel("Nombre")
    ax.legend()
    plt.show

#affichage du graphique 
print(graphe_pop_pourc(130,5000,pi0,A))
```
